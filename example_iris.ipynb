# Empirical Dependence Plot (EDP) — Iris example

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/attilalr/edp-tool/blob/main/example_iris.ipynb)

The EDP shows how a target's **empirical mean** (continuous target) or
**class rate** (binary/multiclass target) changes across grouped values of a
feature. It uses no model and no marginalization — see the README for how it
differs from a true Partial Dependence Plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
import os

url = 'https://raw.githubusercontent.com/attilalr/edp-tool/main/edp_tool.py'
if not os.path.isfile('edp_tool.py'):
    !wget -q {url}

from edp_tool import edp
```

## Load the Iris dataset

```python
data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data.target

features = list(data.feature_names)
df.head()
```

## Multiclass target

The target has three classes. `edp` handles it natively: one line per class,
each showing P(class | feature bin) with a Wilson confidence band. No manual
one-hot encoding needed.

```python
df['species'] = pd.Categorical.from_codes(df['target'], data.target_names)

edp(df, features, 'species', n=4);
```

## Binary target

You can also focus on a single class as a binary 0/1 target.

```python
df['is_virginica'] = (df['target'] == 2).astype(int)

edp(df, features, 'is_virginica', n=4);
```

## Continuous target

With a continuous target, `edp` plots the mean per bin with a standard-error
band. `n` controls the number of bins.

```python
edp(df, ['sepal length (cm)', 'petal length (cm)'], 'petal width (cm)', n=6);
```
